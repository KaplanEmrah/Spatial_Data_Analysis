```python
import pandas as pd
import datetime as dt
import math

def Z_Curve(data):
    max_latitude, max_longitude = max([abs(i[0]) for i in data]), max([abs(i[1]) for i in data])
    max_number_bits = len(bin(int(max([max_latitude, max_longitude]) * 10 ** 7))) - 2 + 1
    c = []
    for value in data:
        a, b = value
        a_ve, b_ve = False, False
        bin_a = bin(int(a*10**7))
        bin_b = bin(int(b*10**7))
        if bin_a[0] == '-':
            a_ve = True
            bin_a = bin_a[3:]
        else:
            bin_a = bin_a[2:]
        if bin_b[0] == '-':
            # print ('here')
            b_ve = True
            bin_b = bin_b[3:]
        else:
            bin_b = bin_b[2:]
        difference1 = max_number_bits - len(bin_a)
        difference2 = max_number_bits - len(bin_b)
        if (difference1 > 0):
            if a_ve == True:
                bin_a = '1' + '0'*(difference1-1) + bin_a
            else:
                bin_a = '0'*(difference1) + bin_a
        if (difference2 > 0):
            if b_ve == True:
                bin_b = '1' + '0'*(difference2-1) + bin_b
            else:
                bin_b = '0'*(difference2) + bin_b
        temp = []
        for i in range(len(bin_a)):
            temp.append(bin_a[i])
            temp.append(bin_b[i])
        temp = ''.join(temp)
        c.append(temp)
    return c

Data = pd.read_csv('user3_gps.csv')
nb_of_data_points = 100000
D = Data[0:nb_of_data_points]
DateTimeData = pd.to_datetime(D['RecordedDate'], dayfirst=True)
Hour = DateTimeData.dt.time

Hour_in_integer = []
for i in range(0, nb_of_data_points):
    temp_H = int(Hour[i].hour)*100
    temp_M = int(Hour[i].minute)
    out = temp_H + temp_M
    Hour_in_integer.append(out)

latitude = [i for i in D['latitude']]
longitude = [i for i in D['longitude']]
Mon = [1 if DateTimeData.dt.dayofweek[i]==0 else 0 for i in range(0, nb_of_data_points)]
Tue = [1 if DateTimeData.dt.dayofweek[i]==1 else 0 for i in range(0, nb_of_data_points)]
Wed = [1 if DateTimeData.dt.dayofweek[i]==2 else 0 for i in range(0, nb_of_data_points)]
Thu = [1 if DateTimeData.dt.dayofweek[i]==3 else 0 for i in range(0, nb_of_data_points)]
Fri = [1 if DateTimeData.dt.dayofweek[i]==4 else 0 for i in range(0, nb_of_data_points)]
Sat = [1 if DateTimeData.dt.dayofweek[i]==5 else 0 for i in range(0, nb_of_data_points)]
Sun = [1 if DateTimeData.dt.dayofweek[i]==6 else 0 for i in range(0, nb_of_data_points)]
Next_Lat = []
Next_Lon = []
for i in range(len(latitude)-1):
    Next_Lat.append(latitude[i+1])
    Next_Lon.append(longitude[i+1])
Next_Lat.append(latitude[0])
Next_Lon.append(longitude[0])

lat_lon = D[['latitude', 'longitude']].as_matrix()
Z = Z_Curve(lat_lon)
Decimal_Z = [int(str(i), 2) for i in Z]

Next_Z = []
Next_Dec_Z = []
for i in range(len(Z)-1):
    Next_Z.append(Z[i+1])
    Next_Dec_Z.append(Decimal_Z[i+1])
Next_Z.append(Z[0])
Next_Dec_Z.append(Decimal_Z[0])

new_dict = {
    'latitude':latitude,
    'longitude':longitude,
    'Z_Value':Z,
    'Decimal_Z_Value': Decimal_Z,
    'Hour':Hour_in_integer,
    'Mon':Mon,
    'Tue':Tue,
    'Wed':Wed,
    'Thu':Thu,
    'Fri':Fri,
    'Sat':Sat,
    'Sun':Sun,
    'Next_Lat':Next_Lat,
    'Next_Lon':Next_Lon,
    'Next_Z':Next_Z,
    'Next_Dec_Z':Next_Dec_Z
}
df_new = pd.DataFrame(new_dict)
df_new.to_csv('user3_gps_DataFrame_Hour_of_Day.csv')
```
